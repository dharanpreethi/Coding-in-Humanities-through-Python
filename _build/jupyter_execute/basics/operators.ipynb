# Operators
An operator is a character that represents an action, as for example x is an arithmetic operator that represents multiplication.

## Arithmetic operators


```python
a=20
b=5
a+b  #addition
```

```python
a/b #division
```

```python
a//b #integer division
```

```python
a-b #subtraction
```

```python
a*b #multiplication
```

```python
a%3  #returns remainder
```

```python
a**2 #power operation 
```
